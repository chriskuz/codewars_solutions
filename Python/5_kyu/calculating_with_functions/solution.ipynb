## Calculating with functions

This time we want to write calculations using functions and get the results. Let's have a look at some examples:

```
seven(times(five())) # must return 35
four(plus(nine())) # must return 13
eight(minus(three())) # must return 5
six(divided_by(two())) # must return 3
```

Requirements:

- There must be a function for each number from 0 ("zero") to 9 ("nine")
- There must be a function for each of the following mathematical operations: plus, minus, times, dividedBy (divided_by in Ruby and Python)
- Each calculation consist of exactly one operation and two numbers
- The most outer function represents the left operand, the most inner function represents the right operand
- Divison should be integer division. For example, this should return 2, not 2.666666...:

`eight(divided_by(three()))`

````python
def zero(*zero_input):
    if zero_input == (): 
        return 0
    else:
        expression  = '0' + zero_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num) 
    
def one(*one_input):
    if one_input == ():
        return 1
    else:
        expression = '1' + one_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num)  
    
def two(*two_input):
    if two_input == ():
        return 2
    else:
        expression = '2' + two_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num)  

def three(*three_input): 
    if three_input == ():
        return 3
    else:
        expression = '3' + three_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num)  

def four(*four_input): 
    if four_input == ():    
        return 4
    else:
        expression = '4' + four_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num) 

def five(*five_input):
    if five_input == ():
        return 5
    else:
        expression = '5' + five_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num)  

def six(*six_input): 
    if six_input == ():
        return 6
    else:
        expression = '6' + six_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num)  

def seven(*seven_input):
    if seven_input == (): 
        return 7
    else:
        expression = '7' + seven_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num) 

def eight(*eight_input): 
    if eight_input == ():
        return 8
    else:
        expression = '8' + eight_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num)  

def nine(*nine_input): 
    if nine_input == ():
        return 9
    else:
        expression = '9' + nine_input[0]

        first_num = int(expression[0])
        second_num = int(expression[2])

        if expression[1] == '+':
            return int(first_num + second_num)
        elif expression[1] == '-':
            return int(first_num - second_num)
        elif expression[1] == '*':
            return int(first_num * second_num)
        elif expression[1] == '/':
            return int(first_num / second_num)  

def plus(plus_input):
    plus_input = str(plus_input)
    return '+' + plus_input    

def minus(minus_input):
    minus_input = str(minus_input)
    return '-' + minus_input

def times(times_input):
    times_input = str(times_input)
    return '*' + times_input

def divided_by(division_input):
    division_input = str(division_input)
    return '/' + division_input
````
